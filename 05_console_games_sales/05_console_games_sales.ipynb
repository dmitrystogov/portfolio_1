## Предобработка данных

### Переименование столбцов

В заголовках столбцов использованы заглавные буквы, это противоречит правилам наименования. Приведем все названия столбцов к строчному регистру.

```python
df.columns = df.columns.str.lower() # переименование столбцов в строчный регистр
df.columns # просмотр наименования столбцов
```

Столбцы успешно переименованы.

### Изменение типов данных

Посмотрим на структуру данных в таблице.

```python
df.info() # получение общей информации о данных в таблице
```

Посмотрим как выглядят данные в таблице.

```python
df.sample(5) # просмотр образцов данных в таблице
```

Данные столбцов `year_of_release` `critic_score` можно привести к типу integer, категории столбца `rating` можно вынести в словарь.

```python
try:
    df['year_of_release'] = df['year_of_release'].astype('Int64') # преобразование в целочисленный тип
except:
    print('Ошибка преобразования типа данных')
df.info() # просмотр результатов преобразования
```

Проверим, есть ли в столбце `critic_score` значения оценки с дробной частью.

```python
df[df['critic_score'] % 1 != 0]['critic_score'].dropna().count() # подсчет значений с дробной частью
```

Все значения оценок целые, можно изменить тип данных столбца.

```python
df['critic_score'] = df['critic_score'].fillna(-111) # Заменим NaN чтобы избежать ошибки
try:
    df['critic_score'] = df['critic_score'].astype('Int64') # преобразование в целочисленный тип
except:
    print('Ошибка преобразования типа данных')
df['critic_score'] = df['critic_score'].replace(-111, np.nan) # вернем NaN обратно
df.info() # просмотр результатов преобразования
```

Тип данных столбца `user_score` object. Посмотрим, какие данные хранятся в этом столбце. 

```python
df['user_score'].sort_values().unique() # просмотр уникальных значений
```

Мы видим, что оценку пользователей можно умножить на 10, это позволит хранить ее как целое, и приведет его по величине к оценке критиков. Строку 'tbd' - "to be determined" для наших целей можно заменить на NaN, у нас узнать будущую оценку нет возможности.

```python
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'] = df['user_score'].fillna(-111) # Заменим NaN чтобы избежать ошибки
try:
    df['user_score'] = df['user_score'].astype('float')
    df['user_score'] = df['user_score'] * 10
    df['user_score'] = df['user_score'].astype('Int64') # преобразование в целочисленный тип
except:
    print('Ошибка преобразования типа данных')
df['user_score'] = df['user_score'].replace(-1110, np.nan) # вернем NaN обратно
df.info() # просмотр результатов преобразования
```

```python
df['user_score'].sort_values().unique()  # просмотр уникальных значений
```

Оценки пользователей теперь целочисленные.

Посмотрим на уникальные значения возрастного рейтинга.

```python
df['rating'].sort_values().unique() # просмотр уникальных значений
```

Создадим словарь со значениями рейтинга.

```python
rate_dic = dict(enumerate(df['rating'].dropna().sort_values().unique().tolist(),  start = 0)) # создание словаря
rate_dic # просмотр словаря
```

Создадим датафрейм, в котором каждому уникальному значению из 'rating' соответствует уникальное значение индекса.

```python
rating_dictionary = pd.DataFrame.from_dict(rate_dic, orient='index', columns = ['rating']) # создание датафрейма из словаря
rating_dictionary # просмотр датафрейма
```

Для замены значений рейтинга в нашей таблице обменяем местами ключи и значения словаря.

```python
rate_dic = dict(zip(rate_dic.values(), rate_dic.keys())) # обмен местами ключей и значений
rate_dic # просмотр словаря
```

Заменим значения рейтинга по словарю.

```python
df['rating'] = df['rating'].replace(rate_dic) # замена значений по словарю
df['rating'].value_counts() # просмотр результатов замены
```

Рейтинг теперь закодирован числами.

Заменим тип данных столбца `rating` на целочисленный.

```python
df['rating'] = df['rating'].fillna(8) # замена NaN На 8, что будет означать'Unknown'
rating_dictionary.loc[8] = ['Unknown'] # добавление этого значения в словарь
try:
    df['rating'] = df['rating'].astype('Int64') # преобразование в целочисленный тип
except:
    print('Ошибка преобразования типа данных')
df.info() # просмотр результатов преобразования
```

Посмотрим на результат замены.

```python
df.sample(5) # просмотр образцов данных
```

Типы данных успешно преобразованы.

Во всех таблицах данные преобразованы к оптимальным типам.

Но, для целей учебного проекта проведем t-тест для проверки гипотезы о равенстве среднего двух генеральных совокупностей.

Проверим критерии применимости t-теста:
* Генеральные совокупности не должны зависеть друг от друга.
* Выборочные средние должны быть нормально распределены.
Критерии удовлетворяются.

Нулевая гипотеза формулируется так, чтобы все данные лежали в районе изначально предполагаемого значения. В нашем случае нулевая гипотеза следующая: 
* средние пользовательские рейтинги платформ Xbox One и PC равны.

Альтернативная :
* средние пользовательские рейтинги платформ Xbox One и PC отличаются.

```python
type(np.array(df[df['platform'] == 'XOne']['user_score'].dropna()))
```

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    np.array(df[df['platform'] == 'XOne']['user_score'].dropna().astype(float)), 
    np.array(df[df['platform'] == 'PC']['user_score'].dropna().astype(float))
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вероятность равенства средних значений выборок пользовательских рейтингов по платформам Xbox One и PC  велика. Принимаем нулевую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC равны. Предложенная заказчиком гипотеза принята.
