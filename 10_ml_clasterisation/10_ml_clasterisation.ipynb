## Описание исследования

**Цель исследования:** Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.

**Исходные данные:** Клиентские анкеты переведенные в электронный вид и информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента. Таблица содержит данные на месяц до оттока и факт оттока на определённый месяц

**Задачи исследования:**

* спрогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
* сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
* проанализировать основные признаки, наиболее сильно влияющие на отток;
* сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:

    1. выделить целевые группы клиентов;
    2. предложить меры по снижению оттока;
    3. определить другие особенности взаимодействия с клиентами.

### Сравнение статистических характеристик групп ушедших и оставшихся

Посмотрим на средние значения признаков в двух группах — тех, кто ушел в отток и тех, кто остался

```python
no_churn_stat = data.query('churn == 0').describe() # рассчитаем статистические характеристики для оставшихся и ушедших 
churn_stat = data.query('churn == 1').describe()
no_churn_stat = no_churn_stat.add_suffix('_no_churn') # добавим к названиям характеристик суффиксы групп 
churn_stat = churn_stat.add_suffix('_churn') # объединим таблицы
pd.concat([no_churn_stat, churn_stat], axis=1).sort_index(axis=1).T # отсортируем по названию характеристик
```

### Построение столбчатых гистограмм и круговых диаграмм признаков

Построим столбчатые гистограммы для тех, кто ушёл (отток) и тех, кто остался (не попали в отток).

Построим гистограммы для небинарных признаков.

```python
# готовим данные для построения графиков
plot_data = data.copy() # скопируем датасет
for columnName in plot_data.columns: # исключим бинарные признаки 
    if (plot_data[columnName].max() == 1) and (columnName != 'churn'):
        plot_data = plot_data.drop(columnName, axis=1)
plot_data.columns # посмотрим получившиеся столбцы 
```

```python
# строим гистограммы
fig, ax = plt.subplots(plot_data.shape[1]-1, 1) 
fig.set_size_inches(5, plot_data.shape[1]*4)
i=0
for columnName in plot_data.columns:
    if (columnName != 'churn'):
        sns.histplot(data = plot_data, x=columnName, kde = False, palette="tab10", hue='churn', ax=ax[0+i]) # построение гистограммы
        i += 1
fig.tight_layout()     
plt.show()
```

Для бинарных признаков построим круговые диаграммы. Они лучше проиллюстрируют соотношения.

## Построение модели прогнозирования оттока клиентов

### Сравнение моделей

Разделим наши данные на признаки (матрица X) и целевую переменную (y).

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = data.drop('churn', axis=1)
y = data['churn']
```

Разделяем данные на обучающую и валидационную выборку.

```python
# разделяем данные на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=0 # учитывая дисбаланс классов применим стратифицирование
)
```

Стандартизируем данные в обучающей и валидационной выборке.

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
X_train_st = scaler.fit_transform(
    X_train
)  # обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
```

```python
# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

Для обучения модели нам предлагается воспользоваться логистической регрессией и случайным лесом.

```python
# зададим список моделей
models = [
    LogisticRegression(solver='liblinear', random_state = 0),
    RandomForestClassifier(n_estimators = 100, random_state = 0)
]
```

Создадим функцию, принимающую на вход модель и данные и выводящую метрики.

```python
# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_true = y_test
    
    # воспользуемся уже обученной моделью, чтобы сделать прогнозы
    y_proba = model.predict_proba(X_test)[:,1]

    # выведем метрики для полученного прогноза
    print('Метрики классификации')
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

Обучим модель на train-выборке, перебирая модели в цикле.

```python
# цикл, который выводит метрики по списку моделей
for i in models:
    print(i)
    make_prediction(i, X_train_st, y_train, X_test_st, y_test)
    print('-'*40)
```

## Кластеризация клиентов

### Построение матрицы расстояний

Отложим в сторону столбец с оттоком и стандартизируем данные.

```python
# стандартизируем данные
sc = StandardScaler()
x_sc = sc.fit_transform(data.drop('churn', axis=1))
```

Построим матрицу расстояний функцией linkage().

```python
# строим матрицу расстояний на стандартизованной матрице признаков 
matrix = linkage(x_sc, method='ward', metric='euclidean') 
```

### Построение дендрограммы

Нарисуем дендрограмму.

```python
# рисуем дендрограмму
plt.figure(figsize=(15, 10))  
dendrogram(matrix, orientation='top', no_labels=True)
plt.title('Дендрограмма кластеров клиентов фитнес-центра')
plt.show()
```

Визуально дендрограмма делится на четыре кластера, но по условиям задания будем делить клиентов на пять кластеров.

### Кластеризация на основании алгоритма K-Means

Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. 

```python
# задаём модель k_means с числом кластеров 5 и фиксируем значение random_state
km =  KMeans(n_clusters=5, random_state=0) 
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 4)
labels = km.fit_predict(x_sc)

# сохраняем метки кластера в поле нашего датасета
data['cluster_km'] = labels
```

Для наглядности сделаем, чтобы номера кластеров показывали среднюю величину оттока.

```python
# отсортируем номера кластеров по средней величине оттока
list_b = data[['cluster_km','churn']].groupby('cluster_km').agg('mean').sort_values('churn', ascending=False).index.to_list()
list_b
```

```python
# заменим номера кластеров
list_a = list(range(0, 5))
data['cluster_km'] = data['cluster_km'].replace(list_b, list_a)
```

Посчитаем количество клиентов в каждом кластере.

```python
#считаем количество клиентов в каждом кластере
data[['cluster_km','churn']].groupby('cluster_km').agg(['count', 'mean'])
```

Кластеры с наибольшим и наименьшим оттоком самые большие, самый маленький - со средним оттоком.

### Построение графиков попарных признаков для кластеров

Определим функцию отрисовки графиков попарных признаков для кластеров.

```python
# определим функцию отрисовки графиков попарных признаков для кластеров
def show_clusters_on_plot(df, x_name, y_name, cluster_name):
    plt.figure(figsize=(5, 5))
    sns.scatterplot(
        df[x_name], df[y_name], hue=df[cluster_name], palette='rocket'
    )
    plt.title('{} vs {}'.format(x_name, y_name))
    plt.show()
```

Отрисуем графики для попарных сочетаний признаков с разметкой по кластерам, исключив бинарные признаки.

```python
# отрисуем графики для попарных сочетаний признаков с разметкой по кластерам
plot_data = data.copy()
for columnName in plot_data.columns:
    if (plot_data[columnName].max() == 1):
        plot_data = plot_data.drop(columnName, axis=1)
col_pairs = list(itertools.combinations(plot_data.drop('cluster_km', axis=1).columns, 2))
for pair in col_pairs:
    show_clusters_on_plot(plot_data, pair[0], pair[1], 'cluster_km')
```

### Рассчет метрики силуэта 

```python
# посчитаем метрику силуэта для нашей кластеризации
print('Silhouette_score: {:.2f}'.format(silhouette_score(x_sc, data['cluster_km'])))
```

Коэффициент силуэта - функция от среднего внутрикластерного расстояния (a) и среднего расстояния до ближайшего кластера (b): (b - a) / max(a, b). Лучшее значение 1, а худшее -1.

Визуально и по метрике Silhouette_score кластеры определились нечетко.

### Расчет средних значения признаков для кластеров
