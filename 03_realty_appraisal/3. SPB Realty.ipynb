## Предобработка данных

### Стиль заголовков
Посмотрим названия столбцов.

```python
df.columns # Посмотрим перечень названий столбцов таблицы df
```

Приведем названия в соответствие с хорошим стилем:

- несколько слов в названии запишем в «змеином_регистре»,
- все символы сделаем строчными,
Для этого переименуем колонки так:

'cityCenters_nearest' → 'city_сenter_nearest';
'parks_around3000' → 'parks_around_3000';
'ponds_around3000' → 'ponds_around_3000'.

Для единообразия переименуем признаки "студия" и "свободная планировка":
'studio' → 'is_studio';
'open_plan' → 'is_open_plan'.

```python
df = (
    df.rename(columns = {'cityCenters_nearest': 'city_сenter_nearest', 'parks_around3000': 'parks_around_3000',
    'studio': 'is_studio', 'open_plan': 'is_open_plan', 'ponds_around3000': 'ponds_around_3000'}) # переименование столбцов
)
df.columns # Выведем перечень названий столбцов таблицы df
```

Столбцы успешно переименованы.

### Пропуски значений
Посчитаем, сколько в таблице пропущенных значений.

```python
df.isna().sum() / len(df) * 100 # Подсчитаем процент пропусков
```

Наибольший процент пропусков в `is_apartment`, `parks_nearest`, `ponds_nearest`.

Изучим возможности заполнения пропусков.

Скорее всего пропуски в столбце `is_apartment` вызваны технической ошибкой. Аналогичные по функции столбцы `is_studio`, `is_open_plan` пропусков не содержат. Посмотрим на содержание этого столбца. Возможно в него вошли только истинные значения.

```python
df['is_apartment'].value_counts() # Подсчитаем количество значений в столбце 'is_apartment'
```

Столбец содержит и истинные и ложные значения, значит какая-то часть данных была введена успешно. Заполнение пропусков ложными значениями приведет к существенным искажениям. Утраты данных в этом столбце слишком велики и при заданных условиях не поддаются восстановлению, поэтому столбец придется удалить.

```python
df = df.drop('is_apartment', axis=1)  # Удалим столбец 'is_apartment'
```

Исследуем столбцы  `parks_nearest`, `ponds_nearest`. Заметно, что в столбцах `parks_around_3000`, `ponds_around3000`, пропусков меньше. Посмотрим, есть ли строки с ненулевыми значениями в столбцах "around_3000" и пропуском в "nearest".

```python
len(df.query('(parks_nearest.isna()) & (parks_around_3000 > 0)')) # Считаем строки с пропуском в parks_nearest и ненулевым parks_around_3000
```

Таких совпадений нет. Посмотрим то же условие для прудов.

```python
len(df.query('(ponds_nearest.isna()) & (ponds_around_3000 > 0)')) # Считаем строки с пропуском в ponds_nearest и ненулевым ponds_around_3000
```

Значит, если у нас указано наличие объекта на расстоянии до 3 км, в этих строках пропусков нет. Учтем это в дальнейшем. Заполнять пропуски в столбцах `parks_nearest`, `ponds_nearest` не будем, поскольку нет возможности подобрать подходящие величины.

То же относится и к столбцам `airports_nearest` и `city_сenter_nearest`. Числа пропусков в них почти одинаковые. Проверим, не пропущены ли они в одних и тех же строках, что подтвердит вероятность технической ошибки.

```python
len(df.query('(airports_nearest.isna()) & (city_сenter_nearest.isna())')) # Считаем строки с пропусками в обоих столбцах
```

Нет, совпадений мало, похоже это ошибка заполняющих данные.

Не будем заполнять столбцы `airports_nearest`, `city_сenter_nearest`, `parks_nearest` и  `ponds_nearest`, поскольку нет возможности подобрать подходящие величины. При оценке влияния факторов строки с пропусками будем просто исключать из рассмотрения. Обратим внимание заказчика, что при автоматическом заполнении таблицы на основе картографических данных происходит много ошибок. 

Посмотрим количество пропусков `ceiling_height`.

```python
df['ceiling_height'].isna().sum() # Посчитаем пропуски в высоте потолка
```

Пропущено больше трети значений. Скорее всего это вызвано небрежностью заполнявших данные, или технической ошибкой. Поскольку на практике разброс значений высоты потолка не велик, заполним пропуски медианным значением.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median()) # Заменим пропуски в высоте потолка медианой
```

```python
df['ceiling_height'].isna().sum() # Посчитаем пропуски в высоте потолка
```

Пропуски устранены. 

Посмотрим количество пропусков в столбце `floors_total`.

```python
df['floors_total'].isna().sum() # Посчитаем пропуски в количестве этажей
```

Количество пропусков невелико. Возможно они вызваны небрежностью заполняющих данные. Проверим есть ли строки в которых указан этаж, но нет общего числа этажей.

```python
len(df.query('(floors_total.isna()) & (floor > 0)')) 
```

Этаж указан во всех случаях. Посмотрим на образец таких строк.

```python
df.query('(floors_total.isna()) & (floor > 0)').sample(5)
```

Заполним пропуски общего числа этажей значениями этажа, на котором находится продаваемая квартира. 

```python
df['floors_total'] = df['floors_total'].fillna(df['floor']) # Заменим пропуски в этажности этажом квартиры
```

Мы видим большое количество выбросов в верхней части диапазона. Для их оценки и устранения создадим функцию для расчета разности и суммы квартилей и полутора межквартильных расстояний. 

```python
def iq_minmax (column, fun):
    '''
    Возвращает в зависимости от параметра либо разность либо сумму первого и третьего
    квартилей и полутора межквартильных расстояний.
    
    Параметры:
    column: столбец для расчета
    fun: выбор функции для расчета:
        'min' - разность первого квартиля и полутора межквартильных расстояний
        'max' - сумма третьего квартиля и полутора межквартильных расстояний
    
    '''
    q25 = column.quantile(0.25)
    q75 = column.quantile(0.75)
    iqr = q75-q25
    if fun == 'min':
        return q25-(1.5*iqr)
    elif fun == 'max':
        return q75+(1.5*iqr)
    else:
        print('Ошибка параметров функции')
```

Посмотрим сколько квартир выходят за рамки суммы третьего квартиля и полутора межквартильных расстояний.

```python
len(df[df['total_area'] > iq_minmax(df['total_area'], 'max')]) # Подсчет числа квартиры с площадью большей суммы 
                                                               # третьего квартиля и 1.5 межквартильных расстояний
```

В Санкт-Петербурге особенностью квартирного фонда является большее число квартир большой площади. Одна из самых больших квартир в Санкт-Петербурге имеет площадь более 800 метров. ### Рекомендации

1. Для разработки автоматизированной системы параметры следует более подробно изучить региональный фактор. Даже выделение центра Санкт-Петербурга показало, что разные параметры влияют на цену квартиры и кв. метра по разному в разных регионах. Возможно следует кластеризировать исследуемую территорию по влиянию параметров.
2. По результатам нашего исследования в качестве параметров для оценки стоимости квартиры можно предложить:
- площадь квартиры;
- отношение площади кухни к общей;
- число комнат;
- высоту потолка;
- расстояние до центра;
- этаж.
3. Оценка зависимости стоимости квартиры от дня недели и месяца публикации возможно пригодна как один из параметров для антифрода.
4. Информация о типичном времени продажи квартиры может быть использована для фильтрации предложений с ценой отличающейся от рыночной, как слишком высокой, так слишком низкой.
5. При дальнейшем исследовании стоимости квартир рекомендуется корректировать данные с учетом инфляции.
