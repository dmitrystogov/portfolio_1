### Сокращение числа событий

Посмотрим уникальные события в логе.

```python
# вывод списка уникальных событий
events['event_name'].sort_values().unique()
```

События `contacts_show` и `show_contacts` относятся к одному действию, назовем их одинаково. События начинающиеся с `search` дают нам слишком мало информации, поэтому объединим их в одно.

### Устранение аномалий

Почти все события имеют выбросы. Подсчитаем перцентили для событий, и определим, будем ли исключать такие выбросы и как.

Рассчитаем перцентили числа событий по их видам.

```python
percentile_list = [90, 95, 99] # перечислим перцентили для расчета
anomaly_margins = []
for col in events_by_users.columns:
    print('Расчет перцентилей для события ', col)
    for i in range(len(percentile_list)): # рассчитаем и выведем перцентили по списку
        print('{} перцентиль = {:1.0f}'.format(percentile_list[i], np.percentile(events_by_users[col], percentile_list)[i]))
    anomaly_margins.append(np.percentile(events_by_users[col], 99))
```

```python
anomaly_margins
```

```python
anomaly_list = []
for col in events_by_users.columns:
    anomaly_list += (events_by_users[events_by_users[col] > anomaly_margins[events_by_users.columns.get_loc(col)]].index.to_list())
anomaly_set = set(anomaly_list)
print('Число аномалий: {}, {:.2%}.'.format(len(anomaly_set), len(anomaly_set)/n_users))
```

### Построение признаков

Создадим матрицу признаков. Сразу включим туда источник, с которого пользователь установил приложение.

```python
features = sourсes.copy()
features = features.set_index('user_id')
```

#### Количество действий в день

Рассчитаем среднее число действий, совершаемых пользователем в день.

```python
# создаем таблицу числа действий пользователей по дням
mean_events = events.pivot_table(index='user_id', columns='event_date', values='event_name', aggfunc='count')
# заменяем нули на nan, чтобы не учитывать дни, когда пользователь не пользовался приложением
mean_events[mean_events == 0] = np.nan
# считаем среднее число действий в день для пользователя
mean_events = mean_events.apply(lambda x: np.nanmean(x), axis=1).to_frame(name='events_per_day')
```

```python
# добавляем признак в таблицу признаков
features = features.merge(mean_events, how='left', left_index=True, right_index=True)
features.head()
```

#### Количество действий в сессию

Рассчитаем среднее число действий, совершаемых пользователем в сессию.

```python
# создаем таблицу числа действий пользователей по сессиям
mean_events = events.pivot_table(index='user_id', columns='session_id', values='event_name', aggfunc='count')
# заменяем нули на nan, чтобы не учитывать дни, когда пользователь не пользовался приложением
mean_events[mean_events == 0] = np.nan
# считаем среднее число действий в день для пользователя
mean_events = mean_events.apply(lambda x: np.nanmean(x), axis=1).to_frame(name='events_per_session')
```

```python
# добавляем признак в таблицу признаков
features = features.merge(mean_events, how='left', left_index=True, right_index=True)
features.head()
```

#### Лайфтайм

Рассчитаем лайфтайм пользователей.

```python
lifetimes = events.groupby('user_id')['event_time'].agg(['min','max'])
lifetimes['lifetime'] = (lifetimes['max'] - lifetimes['min']).dt.days + 1
```

```python
# добавляем признак в таблицу признаков
features = features.merge(lifetimes['lifetime'], left_index=True, right_index=True)
features.head()
```

#### Интервал с первого действия до целевого

Рассчитаем время, которое прошло с момента первого действия пользователя в системе до совершения целевого действия.

```python
# найдем самое раннее время целевого события для пользователя и внесем в таблицу лайфтаймов
lifetimes = lifetimes.merge(events.query('event_name == "contacts_show"')[['user_id', 'event_time']].groupby('user_id').min(), how='left', left_index=True, right_index=True)
```

```python
# посчитаем время от первого до целевого действия
lifetimes['time_to_target'] = (lifetimes['event_time'] - lifetimes['min'])
```

```python
# добавляем признак в таблицу признаков
features = features.merge(lifetimes['time_to_target'], left_index=True, right_index=True)
features.head()
```

Посмотрим, в какое время пользователь совершает первое и последнее действие  в день.

```python
# рассчитаем время начала и конца работы пользователя
events['event_daytime'] = events['event_time'].dt.time
# добавляем признак в таблицу признаков
features = features.merge(events[['user_id', 'event_daytime']].groupby('user_id').agg(['min', 'max']).droplevel(level=0, axis=1)[['min', 'max']], left_index=True, right_index=True)
features = features.rename({'min':'earliest_event','max':'latest_event'}, axis=1)
events = events.drop('event_daytime', axis=1)
features.head()
```

```python
# создаем таблицу числа сессий пользователей по дням
mean_sessions = events.pivot_table(index='user_id', columns='event_date', values='session_id', aggfunc='nunique')
# считаем среднее число действий в день для пользователя
mean_sessions = mean_sessions.apply(lambda x: np.nanmean(x), axis=1).to_frame(name='sessions_per_day')
```

```python
# добавляем признак в таблицу признаков
features = features.merge(mean_sessions, how='left', left_index=True, right_index=True)
features.head()
```

```python
# добавляем признак в таблицу признаков
features = features.merge(tmp, how='left', left_index=True, right_index=True)
features['n_target'] = features['n_target'].fillna(0)
features.head()
```

### Построение матрицы корреляции

Подготовим признаки для построения матрицы корреляции. Для этого заменим категориальные признаки цифровыми, из временных признаков возьмем самую большую величину (день, час)

Скопируем признаки для создания матрицы корреляции.

```python
features_corr = features.copy() # копируем таблицу
```

Создадим словарь для замены категорий источников.

```python
source_dict = dict(enumerate(features_corr['source'].unique())) # создаем словарь уникальных источников
source_dict = dict((v,k) for k,v in source_dict.items()) # меняем местами номера и строки в словаре
source_dict
```

Заменим категории на цифры.

## Рассчет целевых метрик по группам

### Создание функций для расчета удержания и конверсии

Опишем функции для расчета и визуализации удержания и конверсии

#### Функция get_retention

```python
def get_retention(
    profiles, 
    sessions, 
    observation_date, 
    horizon_days, 
    dimensions = [], 
    ignore_horizon = False
):

    '''
    Возвращает таблицы удержания пользователей
    
    Параметры:
    таблицы
    profiles - профили клиентов
    sessions - лог сервера с информацией о посещениях сайта
    переменные
    observation_date - момент анализа
    horizon_days - горизонт анализа данных
    dimensions - "измерение" по которому осуществлять группировку
    ignore_horizon - логический признак игнорировать горизонт анализа
    '''    
    
    
    # оттсекаем всех пользователей, которые не могли "дожить" до нужного горизонта 
    # (присоединились позже чем observation_date - horizon)
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(days = horizon_days - 1)
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # формируем массив сырых данных
    result_raw = result_raw.merge(sessions[['user_id', 'session_start']], on = 'user_id', how = 'left')
    result_raw['lifetime'] = (result_raw['session_start'] - result_raw['first_ts']).dt.days
    
    # функция группировки по нужному набору измерений
    def group_by_dimensions(df, dims, horizon_days):     
        result = df.pivot_table(index = dims, columns = 'lifetime', values = 'user_id', aggfunc = 'nunique')     # строим "треугольную таблицу" 
        cohort_sizes = df.groupby(dims).agg({'user_id': 'nunique'}).rename(columns = {'user_id': 'cohort_size'}) # определяем размеры когорт
        result = cohort_sizes.merge(result, on = dims, how = 'left').fillna(0)                                   # присоединяем размеры когорт к треугольной таблице
        result = result.div(result['cohort_size'], axis = 0)                                                     # делим каждый из столбцов на размер когорты - рассчитываем % удержания (retention rate)
        result = result[['cohort_size'] + list(range(horizon_days))]                                             # оставляем только наблюдения до нужного горизонта и столбец размеров когорт
        result['cohort_size'] = cohort_sizes                                                                     # пересприсваиваем столбец размеров когорт, ведь во время деления он превратился в 1
        return result
    
    # рассчитываем удержание 
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)
    
    # рассчитываем удержание по всем параметрам + дата
    result_in_time = group_by_dimensions(result_raw, dimensions + ['dt'], horizon_days)
    
    # возвращаем и таблицы удержания и сырые данные (чтобы в случае чего можно было в них покопаться при отладке)
    return result_raw, result_grouped, result_in_time
```

#### Функция get_conversion

```python
def get_conversion(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    '''
    Возвращает таблицы конверсии пользователей
    
    Параметры:
    таблицы
    profiles - профили клиентов
    purchases - информация о заказах
    observation_date - момент анализа
    horizon_days - горизонт анализа данных
    dimensions - "измерение" по которому осуществлять группировку
    ignore_horizon - логический признак игнорировать горизонт анализа
    '''       
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # определяем дату и время первой покупки для каждого пользователя
    first_purchases = (
        purchases.sort_values(by=['user_id', 'event_dt'])
        .groupby('user_id')
        .agg({'event_dt': 'first'})
        .reset_index()
    )

    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        first_purchases[['user_id', 'event_dt']], on='user_id', how='left'
    )

    # рассчитываем лайфтайм для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_dt'] - result_raw['first_ts']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу конверсии
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        # считаем сумму с накоплением для каждой строки
        result = result.fillna(0).cumsum(axis = 1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # добавляем размеры когорт в таблицу конверсии
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # делим каждую «ячейку» в строке на размер когорты
        # и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

#### Функция plot_retention

```python
def plot_conversion(conversion, conversion_history, horizon):    
    plt.figure(figsize = (24, 18)) # задаём размер «подложки»

    # исключаем размеры когорт
    # конверсии первого дня различаются, их удалять не нужно
    report = conversion.drop(columns = ['cohort_size'])

    sns.heatmap(
        report, annot=True, fmt='.2%', ax=plt.subplot(2, 1, 1)
    )  # в первой ячейке таблицы графиков строим тепловую карту
    plt.title('Тепловая карта конверсии')

    report.T.plot(
        grid=True, xticks=list(report.columns.values), ax=plt.subplot(2, 2, 3)
    )  # в третьей — кривые конверсии
    plt.title('Кривые конверсии')
    
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in conversion_history.index.names
        if name not in ['dt', 'payer']
    ]
    # для графика истории изменений 
    # преобразуем таблицу динамики конверсии
    report = (
        conversion_history
        .reset_index()
        .pivot_table(index='dt', columns=columns, values=horizon-1, aggfunc='mean')
        .fillna(0)  # заполняем пропуски на случай, если они возникнут
    )
    report.plot(
        # в четвертой ячейке строим график истории изменений
        grid=True, ax=plt.subplot(2, 2, 4)
    )
    plt.title('Динамика конверсии на {}-й день'.format(horizon))
    # исключаем наложение графиков друг на друга
    plt.tight_layout()
    plt.show()   
```

Применим z-тест к долям пользователей, совершивших целевое событие. 

Для сравнения групп создадим функцию проводящую z-тест.

```python
def z_test(name_list, sample_1, sample_2, alpha):
    '''
    Выводит на экран результат z-теста и промежуточные значения
    Параметры:
    name_list - список названий групп
    sample_1, sample_2 - выборки для тестирования
    alpha - критический уровень статистической значимости если p-value окажется меньше него - отвергнем гипотезу
    '''
    # число пользователей и успехов в первой группе:
    n_users_1 = sample_1.count()
    n_actors_1 = sample_1[sample_1 != 0].count()
    
    # число пользователей и успехов во второй группе:
    n_users_2 = sample_2.count()
    n_actors_2 = sample_2[sample_2 != 0].count()
    

    # пропорция успехов в первой группе:
    p1 = n_actors_1/n_users_1

    # пропорция успехов во второй группе:
    p2 = n_actors_2/n_users_2

    # пропорция успехов в комбинированном датасете:
    p_combined = (n_actors_1 + n_actors_2) / (n_users_1 + n_users_2)

    # разница пропорций в датасетах
    difference = p1 - p2  
    
    # вывод промежуточных значений
    print('Сравнение групп {} и {} по целевому событию'.format(name_list[0], name_list[1]))
    print('Число пользователей совершивших целевое событие в группе {}: {}'.format(name_list[0], n_actors_1))
    print('Доля от общего числа пользователей: {:0.2%}'.format(p1))
    print('Число пользователей совершивших целевое событие в группе {}: {}'.format(name_list[1], n_actors_2))
    print('Доля от общего числа пользователей: {:0.2%}'.format(p2))
    
    # считаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / math.sqrt(
        p_combined * (1 - p_combined) * (1 / n_users_1 + 1 / n_users_1))
        
    # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = st.norm(0, 1)

    p_value = (1 - distr.cdf(abs(z_value))) * 2

    # вывод результата тестирования
    print('Проверим нулевую гипотезу: между долями пользователей совершивших действие в разных группах нет значимой разницы')
    print('критический уровень статистической значимости', alpha)
    print('p-значение: ', p_value)
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print(
            'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
        )
        
        
        
    # тест односторонний вправо:
    # отвергаем гипотезу только тогда, когда доля в группе A значимо больше, чем доля в группе B
    
    #if (p_value / 2 < alpha) and (p2 > p1):
    #    print("Отвергаем нулевую гипотезу")
    #else:
    #    print("Не отвергаем нулевую гипотезу") 
    
```

Опишем функцию для разбиения на группы по времени совершения целевого события.

```python
def group_by_days(time):
    if time != time:
        group = 0
    elif time <= pd.Timedelta("1 days"):
        group = 1
    elif time >= pd.Timedelta("7 days"):
        group = 2
    else:
        group = 0
    return group
```

```python
sample = features.join(sample.set_index('user_id')['group'])[['group','n_target']]
sample.head()
```

Для проверки гипотезы воспользуемся t-критерием Стьюдента.

Будем проверять нулевую гипотезу: средние выборок, совершивших целевое действие в первый день и совершивших целевое действие позже 7 дней, равны, соответственно альтернативная гипотеза: между средними этих выборок есть значимая разница и среднее первой группы больше среднего второй. Критический уровень статистической значимости установим в 5%.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    sample[sample['group'] == 1]['n_target'], 
    sample[sample['group'] == 2]['n_target'], equal_var = False)

print('дисперсия выборки 1:', sample[sample['group'] == 1]['n_target'].var())
print('дисперсия выборки 2:', sample[sample['group'] == 2]['n_target'].var())
print('среднее выборки 1:', sample[sample['group'] == 1]['n_target'].mean())
print('среднее выборки 2:', sample[sample['group'] == 2]['n_target'].mean())

# тест односторонний: p-value будет в два раза меньше
print('p-значение:', results.pvalue / 2)

# тест односторонний влево:
# отвергаем гипотезу только тогда, когда выборочное среднее значимо меньше предполагаемого значения
if (results.pvalue / 2 < alpha) and (sample[sample['group'] == 1]['n_target'].mean() < sample[sample['group'] == 2]['n_target'].mean()):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу") 
```

Не получилось отвергнуть нулевую гипотезу. Между средними двух групп нет статистической разницы.
